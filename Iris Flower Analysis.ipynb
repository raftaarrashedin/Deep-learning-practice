Wonder what species of IRIS do these flowers belong to?

Import the Libraries

```python
# Import the Libraries

# Loading the Library with the iris dataset
from sklearn.datasets import load_iris

# Loading scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier

#Loading Pandas & Numpy
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
#Setting random seed 
np.random.seed(0)
```

```python
# Creating an Object called iris with iris dataset
# iris = load_iris()
# df = pd.DataFrame(iris.data, columns= iris.feature_names)
# df.head()

#from dataset file of iris.csv
iris = pd.read_csv('Iris.csv')
# Displaying first 5 rows of our dataset
iris.head()
```

```python
# Displaying last five rows of our dataset
iris.tail()
```

```python
# Now we gonna add one column

# Creating Test and Train Data
iris['is_train'] = np.random.uniform(0,1, len(iris)) <=.75

#Displaying the top 5 rows
iris.head()
```

```python
# Creating the data frames with test rows and training rows
train, test = iris[iris['is_train'] == True], iris[iris['is_train'] == False]

# Show the number of observation for the test and training data frames
print('Number of observation in the training dataset : ', len(train))
print('Number of observation in the test dataset : ', len(test))
```

```python
# Creating a list of the Feature column's names
features = iris.columns[1:5]

# View Features
print(features )
```

```python
# Converting each species name into digits
y = pd.factorize(train['Species'])[0]

#viewing target
y
```

```python
# Creating a Random Forest Classifier
clf = RandomForestClassifier(n_jobs = 2, random_state = 0)

# Training the classifier
print(clf.fit(train[features], y))
```

```python
# Applying the trained classifier to the test
test[features]
clf.predict(test[features])
```

```python
# Viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[10:20]
```

```python
# Mapping names for the plants for each predicted plant class
preds = iris.Species[clf.predict(test[features])]


# View the PREDICTED species for the first five observations
preds[0:5]
```

```python
# Viewing the Actual species for the first five observations
test['Species'].head()
```
