```python
def perceptron(X,y,b) :
  w1 = w2 = b = 1
  learning_rate = 0.1

  for j in range(1000) :
    for i in range(X.shape[0]) :
      
      # check condition
      z = w1 * X[i][0] + w2 * X[i][1] + b

      if z*y[i] < 0 :
        w1 = w1 + learning_rate * y[i] * X[i][0]
        w2 = w2 + learning_rate * y[i] * X[i][1]
        b = b + learning_rate * y[i]

  
  return
```

```python
w1,w2,b = perceptron(X,y,b)
```

```python
m = -(w1/w2)
c = -(b/w2)
print(m,c)
```
