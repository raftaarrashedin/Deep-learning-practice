```python
def initialize_parameters(layer_dims) :
  np.random.seed(3)
  parameter = {}
  l = len(layer_dims)

  for l in range(1,l) :
    parameter['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameter['b' + str(l)] = np.zeros((layer_dims[l],1))

  return parameter
```

```python

def linear_forward(a_prev, w, b) :
  
  z = np.dot(w.T,a_prev) + b

  return z
```
